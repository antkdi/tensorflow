```python
## ... 다음 사람이 한다
## pip install matplotlib 선행
## pip install pillow << 설치하지 않으면 png 만 사용한다.

import matplotlib.image as mp_image
filename = "str.jpg"
input_image = mp_image.imread(filename)

print ('input dim = {}'.format(input_image.ndim))
print ('input shape = {}'.format(input_image.shape))
```

```python
## 이미지 불러오기
import matplotlib.pyplot as plt
plt.imshow(input_image)
plt.show()
```

```python
## 이미지 자르기
## placeholder 를 이용해 잘라진 이미지를 저장할 수 있는 변수를 생성해야함
import tensorflow as tf
my_image = tf.placeholder("uint8", [None,None,3])

## 자르기 연산은 slice 사용
slice = tf.slice(my_image,[10,0,0],[16,-1,-1])

with tf.Session() as session:
    result = session.run(slice,feed_dict={my_image: input_image})
    print(result.shape)
plt.imshow(result) ## 이미지를 불러보자!
plt.show()
```

```python
## 전치연산을 사용하여 이미지를 전치한다.
import tensorflow as tf
x = tf.Variable(input_image,name='x') ## 입력 이미지는 변수 x에 저장
## model=tf.initialize_all_variables() ## 모델 초기화, 2017-03-02 이후로 deprecated 됨
model=tf.global_variables_initializer()

with tf.Session() as session:
    x=tf.transpose(x,perm=[1,0,2]) ## transpose 는 0,1번의 입력행렬 축을 각각 바꾸며 z축은 유지
    session.run(model)
    result=session.run(x)
plt.imshow(result)
plt.show()
```
